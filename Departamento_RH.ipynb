```python
# criando variáveis funcionários que saíram e funionários que ficaram
left_df = employee_df[employee_df['Attrition'] == 1]
stayed_df = employee_df [employee_df['Attrition'] == 0]
```

```python
# Analisando quantidade e percentual de funcionários que saíram ou permaneceram na empresa
print('Total = ', len(employee_df))
print('Quantidade de Funcionários que saíram da empresa = ', len(left_df))
print('Porcentagem de Funcionários que saíram da empresa = ', (len(left_df) / len(employee_df)) * 100)
print('Quantidade de Funcionários que permaneceram na empresa = ', len(stayed_df))
print('Porcentagem de Funcionários que permeneceram na empresa = ', (len(stayed_df) / len(employee_df)) * 100)
```

#### KDE (Kernel Density Estimate) - Gráfico para visualizar a densidade de probabilidade de uma variável numérica

```python
#KDE (Kernel Density Estimate) - Gráfico para visualizar a densidade de probabilidade de uma variável numérica
# parâmetro shade para o preeenchimento
plt.figure(figsize = (12,7))
sns.kdeplot(left_df['DistanceFromHome'], label = 'Funionários que sairam',shade=True, color='r');
sns.kdeplot(stayed_df['DistanceFromHome'], label = 'Funionários que não sairam',shade=True, color='b');

# conclusão - a densidade dos funcionários que ficam é maior entre 2 e 3 km, 
# e a proporção da pessoas que ficaram com mais de 10 km é menor do que a de pessoas que sairam
```

```python
plt.figure(figsize = (12,7))
sns.kdeplot(left_df['TotalWorkingYears'], label = 'Funionários que sairam',shade=True, color='r');
sns.kdeplot(stayed_df['TotalWorkingYears'], label = 'Funionários que não sairam',shade=True, color='b');

# conclusão - a densidade das pessoas que ficaram é maior do que a de pessoas que saíram, 
#indica que as pessoas que ficam foram as que trabalharam mais tempo ao longo da vida
```

```python
# Gerando um Boxplot (gráfico de caixa)
# verificar se tem algum tipo de ligação entre o salário e o gênero
# Primeiro Quartil - 25% da base de dados 
# Segundo Quartil (Mediana) - 50% da base de dados
# Terceiro Quartil - 75% da base de dados
plt.figure(figsize =(15,10));
sns.boxplot(x= 'MonthlyIncome', y = 'Gender', data = employee_df);

# conclusão - podemos verificar que a mediana para o sexo feminino é maior do que a masculina, 
# indicando que a mediana das funcionárias mulheres ganham mais que os homens
# Conclusão -  em relação aos outliers, salários acima da máxima, o gênero masculino são maiores do que o gênero feminino
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x = 'MonthlyIncome', y='JobRole', data = employee_df);

# conclusão - confirma-se o que foi constatado anteriormente que o cargo menor(Sales Representative) remunerado é o que tem mais evasão
```
